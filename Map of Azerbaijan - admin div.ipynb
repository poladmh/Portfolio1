```python
saylist = list(range(0,79))
adlist = []
for sayi in saylist:
    adlist.append(aze_rayon["features"][sayi]["properties"]["name_2"])
# geojson fayldan rayon adlarini goturub fake statistika dataframe hazirlanmasinda istifade etdim
# oxunus ucun cetin olsa da qarsilasdirma ucun uygun property
```

```python
saylist2 = list(range(0,79))
adlist2 = []
for sayi in saylist2:
    adlist2.append(aze_rayon["features"][sayi]["properties"]["varname_2"])
# oxunus ucun uygun propertyni de ayrica goturdum ki xeritede daha rahat oxunsun
```

```python
dfdict={"Name":adlist,"ID":list(range(1,80)),"Population":random.sample(range(10000, 1000000), 79)
        ,"Density":random.sample(range(1,100),79),"Ad":adlist2}
df = pd.DataFrame(dfdict)
# numune ucun fake statistika ehtiva eden dataframe hazirladim
```

```python
geodict = {}
for feature in aze_rayon["features"]:
    feature["id"] = feature["properties"]["id_2"]
    geodict[feature["properties"]["name_2"]] = feature["id"]

# hazirladigim dataframe ile geojson faylindaki rayon adlarini qarsilasdiraraq ID eynilesdirmesi etdim
```

```python
df["GeoID"]=df["Name"].apply(lambda x:geodict[x])
#eynilesdirilmis ID'leri dataframe elave etdim
```

```python
fig = px.choropleth_mapbox( data_frame=df
                           ,locations = "GeoID"
                           ,geojson=aze_rayon
                           ,color="Population"
                           ,mapbox_style="white-bg"
                           ,center={"lat":40.1,"lon":48}
                           ,zoom = 5.9
                           ,opacity=0.7
                           ,hover_name ="Ad"
                           ,hover_data=["Density"]
                           ,title="Azərbaycan rayonlarının əhali sayı"
                           ,)
fig.show()
#plotly kitabxanasindan istifade ederek geojson fayli esasinda fakedataframe'i vizuallasdirdim
```
